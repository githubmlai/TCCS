```python
%matplotlib inline
from __future__ import division  # force floating point division

from IPython.html.widgets import interact, interactive
from IPython.html import widgets
from IPython.display import clear_output, display, HTML
import numpy as np
from matplotlib import pyplot as plt

import model_generation as mg

def plot_von_karman_energy_spectrum(num_sample, 
                                hurst_related_exponent,
                                characteristic_dist):
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    list_wave_number = np.arange(num_sample/2) / num_sample
    von_karman_energy_spectrum = mg.calc_von_karman_energy_spectrum(list_wave_number,
                                                                    hurst_related_exponent,
                                                                    characteristic_dist)
    plt.plot(list_wave_number,von_karman_energy_spectrum)
    plt.show()
    return von_karman_energy_spectrum
    
list_wave_number_in = np.arange(num_sample/2) / num_sample
plot_von_karman_energy_spectrum(num_sample = 2**5, 
                                hurst_related_exponent = 0.25,
                                characteristic_dist = 3)

w = interactive(plot_von_karman_energy_spectrum,
             num_sample=(2**5,2**10),
             hurst_related_exponent=(-0.25,1),
             characteristic_dist=(0.1,10))
display(w)
```
